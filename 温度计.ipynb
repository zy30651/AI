```python
import matplotlib.pyplot as plt
import random
# 画出温度变化图
# 创建figure
plt.figure(figsize=(10, 4), dpi=80)
# 准备x, y坐标的数据
x = range(60)
y_bj = [random.uniform(15, 18) for i in x]
# 画折线图
plt.plot(x, y_bj, label="北京")

x_ch = ["11点{}分".format(i) for i in x]
y_ticks = range(40)
# 修改刻度值
# 指定显示的x刻度的列表
plt.xticks(x[::5], x_ch[::5], fontproperties='SimHei')
plt.yticks(y_ticks[::5], fontproperties='SimHei')

# 增加标题， x轴，y轴描述信息
plt.xlabel("时间", fontproperties='SimHei')
plt.ylabel("温度", fontproperties='SimHei')
plt.title("中午11点0分到12点之间北京的温度变化图")
plt.show()
```
